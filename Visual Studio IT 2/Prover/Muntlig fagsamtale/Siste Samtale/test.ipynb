```python
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        if df[col].max() > 2**63 - 1 or df[col].min() < -2**63:
            print(col)
```

```python
for col in df.columns:
    if df[col].dtype == 'int64':
        df[col] = df[col].astype('int8')
    elif df[col].dtype == 'float64':
        df[col] = df[col].astype('float16')

df.info(memory_usage='deep')
```

```python
forwards = ["SF", "PF"]
guards = ["PG", "SG"]

guards_mean = df[df["Pos"].isin(guards)]["PTS"].mean()
forwards_mean = df[df["Pos"].isin(forwards)]["PTS"].mean()

guards_std = df[df["Pos"].isin(guards)]["PTS"].std()
forwards_std = df[df["Pos"].isin(forwards)]["PTS"].std()

print(f"Forwards: {forwards_mean} mean PTS, Guards: {guards_mean} mean PTS")
print(f"Forwards: {forwards_std} std PTS, Guards: {guards_std} std PTS")
```
